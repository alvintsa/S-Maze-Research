```python
# importing all necessary modules
from gensim.models import Word2Vec
import gensim
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
import string
```

```python
def tokenize_text(text):
    preprocessed_text = text.replace("\n", " ")
    
    data = []
    # iterate through each sentence in the file
    for i in sent_tokenize(preprocessed_text):
        temp = []
    
        # tokenize the sentence into words
        for j in word_tokenize(i):
            temp.append(j.lower())
    
        data.append(temp)
    
    return data
```

```python
text = "She is a cool person"
tokens = tokenize_text(text)
model1 = gensim.models.Word2Vec(tokens, min_count=1,
								vector_size=100, window=5, sg = 0)

vocab = list(model1.wv.index_to_key)  # `index_to_key` contains words in order of frequency
vocab
```

```python
# Python program to generate word vectors using Word2Vec

warnings.filterwarnings(action='ignore')

text = open("./data/alice.txt").read()
data = tokenize_text(text)

# Create CBOW model
model1 = gensim.models.Word2Vec(data, min_count=1,
								vector_size=100, window=5)

# Print results
print("Cosine similarity between 'alice' " +
	"and 'wonderland' - CBOW : ",
	model1.wv.similarity('alice', 'wonderland'))

print("Cosine similarity between 'alice' " +
	"and 'machines' - CBOW : ",
	model1.wv.similarity('alice', 'machines'))

# Create Skip Gram model
model2 = gensim.models.Word2Vec(data, min_count=1, vector_size=100,
								window=5, sg=1)

# Print results
print("Cosine similarity between 'alice' " +
	"and 'wonderland' - Skip Gram : ",
	model2.wv.similarity('alice', 'wonderland'))

print("Cosine similarity between 'alice' " +
	"and 'machines' - Skip Gram : ",
	model2.wv.similarity('alice', 'machines'))
```

```python
alice_vocab = list(model1.wv.index_to_key)  # `index_to_key` contains words in order of frequency
alice_vocab
```
